# Visualization For Data Analysis

# Boxplot of Magnitude column

 A boxplot is a graph that gives a visual indication of how a data set's 25th percentile, 50th percentile, 75th percentile, minimum, maximum and outlier values are spread out and compare to each other.

```python
sns.boxplot(x=s['Magnitude'],palette='hls')
sns.despine()
plt.show()
```

```python
q1=s["Magnitude"].quantile(0.25)
q3=s["Magnitude"].quantile(0.75)
iqr=q3-q1
lower_limit = q1 - 1.5*iqr

upper_limit = q3 + 1.5*iqr

lower_limitoutliers = s[s['Magnitude'] < lower_limit]

upper_limitoutliers = s[s['Magnitude'] > upper_limit]
new_df= s[(s['Magnitude'] < upper_limit)&(s['Magnitude'] >lower_limit)]

import matplotlib.pyplot as plt

import seaborn as sns

sns.set(rc={"figure.figsize": (6, 6)})

sns.boxplot(x=new_df["Magnitude"])

plt.legend(loc="best")
```
